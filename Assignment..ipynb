```python
arr = [21, 10, 45, 32, 67, 5, 18]
def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1

print("Index of 32:", linear_search(arr, 32))
```

```python
arr = [21, 10, 45, 32, 67, 5, 18]
def binary_search(arr, target):
    low = 0
    high = len(arr) - 1
    
    while low <= high:
        mid = (low + high) // 2
        
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
            
    return -1

sorted_arr = sorted(arr)
print("Sorted array:", sorted_arr)
print("Index of 45:", binary_search(sorted_arr, 45))
```

```python
arr = [21, 10, 45, 32, 67, 5, 18]
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr

print("Sorted array:", bubble_sort(arr.copy()))
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
```

```python
class LinkedList:
    def __init__(self):
        self.head = None

    # Insert at beginning
    def insert_beginning(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    # Insert at end
    def insert_end(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        
        current = self.head
        while current.next:
            current = current.next
        current.next = new_node

    # Display elements
    def display(self):
        current = self.head
        while current:
            print(current.data, end=" -> ")
            current = current.next
        print("None")
```

QUESTION 5

The most appropriate data structure is a Queue.

Justification:
A queue follows the FIFO (First In, First Out) principle.
The first student to arrive is the first to be served — just like in a real finance office line.

```python
from collections import deque

queue = deque()

# Add student
def add_student(name):
    queue.append(name)
    print(name, "added to the queue.")

# Serve student
def serve_student():
    if queue:
        served = queue.popleft()
        print(served, "has been served.")
    else:
        print("No students in the queue.")

# Display queue
def display_queue():
    print("Current queue:", list(queue))


# Example usage
add_student("Alice")
add_student("Brian")
display_queue()
serve_student()
display_queue()
```
