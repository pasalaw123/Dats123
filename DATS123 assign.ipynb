```python
# Node class
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


# Linked List class
class SinglyLinkedList:
    def __init__(self):
        self.head = None

    # Print all elements
    def display(self):
        current = self.head
        while current:
            print(current.data, end=" -> ")
            current = current.next
        print("None")

    # Insert at beginning
    def insert_beginning(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    # Insert at end
    def insert_end(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return

        current = self.head
        while current.next:
            current = current.next
        current.next = new_node

    # Insert at specific position (position starts from 1)
    def insert_position(self, data, position):
        new_node = Node(data)

        if position == 1:
            new_node.next = self.head
            self.head = new_node
            return

        current = self.head
        for i in range(position - 2):
            if not current:
                print("Position out of range")
                return
            current = current.next

        new_node.next = current.next
        current.next = new_node

    # Delete first node
    def delete_first(self):
        if self.head:
            self.head = self.head.next

    # Delete last node
    def delete_last(self):
        if not self.head:
            return

        if not self.head.next:
            self.head = None
            return

        current = self.head
        while current.next.next:
            current = current.next
        current.next = None

    # Search element
    def search(self, key):
        current = self.head
        position = 1
        while current:
            if current.data == key:
                return position
            current = current.next
            position += 1
        return -1

    # Count nodes
    def count_nodes(self):
        count = 0
        current = self.head
        while current:
            count += 1
            current = current.next
        return count

    # Reverse list
    def reverse(self):
        prev = None
        current = self.head

        while current:
            next_node = current.next
            current.next = prev
            prev = current
            current = next_node

        self.head = prev


# -----------------------------
# Testing All Operations
# -----------------------------

sll = SinglyLinkedList()

# i. Create list with 3 nodes
sll.insert_end(10)
sll.insert_end(20)
sll.insert_end(30)

print("Initial List:")
sll.display()

# ii. Insert at beginning
sll.insert_beginning(5)
print("After inserting at beginning:")
sll.display()

# iii. Insert at end
sll.insert_end(40)
print("After inserting at end:")
sll.display()

# iv. Insert at position 2
sll.insert_position(15, 2)
print("After inserting 15 at position 2:")
sll.display()

# v. Delete first node
sll.delete_first()
print("After deleting first node:")
sll.display()

# vi. Delete last node
sll.delete_last()
print("After deleting last node:")
sll.display()

# vii. Search for element
element = 20
position = sll.search(element)
if position != -1:
    print(f"Element {element} found at position {position}")
else:
    print("Element not found")

# viii. Count nodes
print("Total number of nodes:", sll.count_nodes())

# ix. Reverse the list
sll.reverse()
print("After reversing the list:")
sll.display()
```

```python
# Node class for Doubly Linked List
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None


# Doubly Linked List class
class DoublyLinkedList:
    def __init__(self):
        self.head = None

    # i. Display forward
    def display_forward(self):
        temp = self.head
        while temp:
            print(temp.data, end=" <-> ")
            temp = temp.next
        print("None")

    # i. Display backward
    def display_backward(self):
        temp = self.head
        if temp is None:
            return

        while temp.next:
            temp = temp.next

        while temp:
            print(temp.data, end=" <-> ")
            temp = temp.prev
        print("None")

    # ii. Insert at beginning
    def insert_beginning(self, data):
        new_node = Node(data)
        if self.head is not None:
            self.head.prev = new_node
            new_node.next = self.head
        self.head = new_node

    # ii. Insert at end
    def insert_end(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            return

        temp = self.head
        while temp.next:
            temp = temp.next

        temp.next = new_node
        new_node.prev = temp

    # iii. Insert at position (position starts at 1)
    def insert_position(self, data, position):
        new_node = Node(data)

        if position == 1:
            self.insert_beginning(data)
            return

        temp = self.head
        for i in range(position - 2):
            if temp is None:
                print("Position out of range")
                return
            temp = temp.next

        if temp.next is None:
            self.insert_end(data)
        else:
            new_node.next = temp.next
            new_node.prev = temp
            temp.next.prev = new_node
            temp.next = new_node

    # iv. Delete from beginning
    def delete_beginning(self):
        if self.head is None:
            return

        if self.head.next is None:
            self.head = None
        else:
            self.head = self.head.next
            self.head.prev = None

    # iv. Delete from end
    def delete_end(self):
        if self.head is None:
            return

        temp = self.head
        if temp.next is None:
            self.head = None
            return

        while temp.next:
            temp = temp.next

        temp.prev.next = None

    # v. Delete from position
    def delete_position(self, position):
        if self.head is None:
            return

        if position == 1:
            self.delete_beginning()
            return

        temp = self.head
        for i in range(position - 1):
            if temp is None:
                print("Position out of range")
                return
            temp = temp.next

        if temp.next is None:
            temp.prev.next = None
        else:
            temp.prev.next = temp.next
            temp.next.prev = temp.prev

    # vi. Search element
    def search(self, key):
        temp = self.head
        position = 1
        while temp:
            if temp.data == key:
                return position
            temp = temp.next
            position += 1
        return -1

    # vii. Count nodes
    def count_nodes(self):
        count = 0
        temp = self.head
        while temp:
            count += 1
            temp = temp.next
        return count

    # viii. Reverse list by changing links
    def reverse(self):
        temp = None
        current = self.head

        while current:
            temp = current.prev
            current.prev = current.next
            current.next = temp
            current = current.prev

        if temp:
            self.head = temp.prev


# -----------------------------
# Testing All Operations
# -----------------------------

dll = DoublyLinkedList()

# i. Create list with 3 nodes
dll.insert_end(10)
dll.insert_end(20)
dll.insert_end(30)

print("Forward:")
dll.display_forward()

print("Backward:")
dll.display_backward()

# ii. Insert at beginning and end
dll.insert_beginning(5)
dll.insert_end(40)
print("After inserting at beginning and end:")
dll.display_forward()

# iii. Insert at position 3
dll.insert_position(15, 3)
print("After inserting 15 at position 3:")
dll.display_forward()

# iv. Delete from beginning and end
dll.delete_beginning()
dll.delete_end()
print("After deleting beginning and end:")
dll.display_forward()

# v. Delete from position 2
dll.delete_position(2)
print("After deleting position 2:")
dll.display_forward()

# vi. Search element
pos = dll.search(20)
if pos != -1:
    print("Element found at position:", pos)
else:
    print("Element not found")

# vii. Count nodes
print("Total nodes:", dll.count_nodes())

# viii. Reverse list
dll.reverse()
print("After reversing:")
dll.display_forward()
```

```python
# Node class
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


# Circular Singly Linked List class
class CircularSLL:
    def __init__(self):
        self.head = None

    # i. Create & traverse once
    def display_once(self):
        if self.head is None:
            print("List is empty")
            return

        temp = self.head
        while True:
            print(temp.data, end=" -> ")
            temp = temp.next
            if temp == self.head:
                break
        print("(back to head)")

    # ii. Insert at beginning
    def insert_beginning(self, data):
        new_node = Node(data)

        if self.head is None:
            self.head = new_node
            new_node.next = new_node
            return

        temp = self.head
        while temp.next != self.head:
            temp = temp.next

        new_node.next = self.head
        temp.next = new_node
        self.head = new_node

    # ii. Insert at end
    def insert_end(self, data):
        new_node = Node(data)

        if self.head is None:
            self.head = new_node
            new_node.next = new_node
            return

        temp = self.head
        while temp.next != self.head:
            temp = temp.next

        temp.next = new_node
        new_node.next = self.head

    # iii. Delete first node
    def delete_first(self):
        if self.head is None:
            return

        if self.head.next == self.head:
            self.head = None
            return

        temp = self.head
        while temp.next != self.head:
            temp = temp.next

        self.head = self.head.next
        temp.next = self.head

    # iii. Delete last node
    def delete_last(self):
        if self.head is None:
            return

        if self.head.next == self.head:
            self.head = None
            return

        temp = self.head
        prev = None
        while temp.next != self.head:
            prev = temp
            temp = temp.next

        prev.next = self.head

    # iv. Search element
    def search(self, key):
        if self.head is None:
            return -1

        temp = self.head
        position = 1

        while True:
            if temp.data == key:
                return position
            temp = temp.next
            if temp == self.head:
                break
            position += 1

        return -1

    # v. Count nodes
    def count_nodes(self):
        if self.head is None:
            return 0

        count = 0
        temp = self.head
        while True:
            count += 1
            temp = temp.next
            if temp == self.head:
                break

        return count

    # vi. Traverse N times
    def traverse_n_times(self, n):
        if self.head is None:
            print("List is empty")
            return

        temp = self.head
        total_nodes = self.count_nodes()
        steps = total_nodes * n

        for i in range(steps):
            print(temp.data, end=" -> ")
            temp = temp.next

        print("(stopped after", n, "cycles)")
        

# -----------------------------
# Testing All Operations
# -----------------------------

csll = CircularSLL()

# i. Create list with 3 nodes
csll.insert_end(10)
csll.insert_end(20)
csll.insert_end(30)

print("Initial Circular List (once):")
csll.display_once()

# ii. Insert at beginning and end
csll.insert_beginning(5)
csll.insert_end(40)
print("After inserting at beginning and end:")
csll.display_once()

# iii. Delete first and last node
csll.delete_first()
csll.delete_last()
print("After deleting first and last:")
csll.display_once()

# iv. Search element
pos = csll.search(20)
if pos != -1:
    print("Element found at position:", pos)
else:
    print("Element not found")

# v. Count nodes
print("Total nodes:", csll.count_nodes())

# vi. Traverse 2 times
print("Traversing 2 times:")
csll.traverse_n_times(2)
```
